## Look up scattering factors in table

```python
import numpy as np
# Scattering factors for QSTEM come from Rez et al., Acta Cryst A 50 (1994) 481
scatFactors = np.loadtxt("SF_rez.txt",delimiter=",")
# There are 93 rows with 123 scattering factors, each.  These are calculated
#   from 0 to 6 A
print scatFactors.shape
# scatFactors[0] is the radius for the scattering factor
# every other row is the corresponding element.  I.e. Carbon = scatFactors[6]
```

```python
# Plot the scattering factors we've loaded
# log of scattering factor is shown
plt.plot(scatFactors[0],scatFactors[6],color="blue", label="C")
plt.plot(scatFactors[0],scatFactors[14],color="red", label="Si")
plt.plot(scatFactors[0],scatFactors[29],color="black", label="Cu")
plt.plot(scatFactors[0],scatFactors[79],color="purple", label="Au")
plt.plot(scatFactors[0],scatFactors[92],color="green", label="U")

plt.title("Scattering factors: C, Si, Cu, Au, U")
plt.xlabel("Scattering Angle, 1/Angstrom")
plt.ylabel("Scattering Factor, Angstrom")
plt.legend()
# show only the first 2 1/A scattering angles, since they taper off to 0.
plt.axis((0,2,0,scatFactors[92].max()))
```

```python
# Here we interpolate the tabulated scattering factors as necessary.

from scipy import interpolate

# store potential boxes in a dictionary for later addition to sample box
potentials={}

def getAtomPotentialBox(atomBox, Z):
    radius=atomBox/2.0 # angstroms from atom center
    # establish our number of points as the atom box divided by our 
    #    (real-space) resolution
    pts = atomBox/d
    # setup cubic spline interpolation for potential (Sr atom):
    tck = interpolate.splrep(scatFactors[0],scatFactors[38],s=0)
    axis = np.linspace(-radius,radius,pts)

    spaceX, spaceY, spaceZ = np.meshgrid(axis,axis,axis)
    # shape 3D space into 1D for interpolation
    space = (spaceX**2+spaceY**2+spaceZ**2).reshape(-1)
    potential = interpolate.splev(np.sqrt(space),tck,der=0)
    # reshape back to 3D
    potential = potential.reshape((pts,pts,pts))
    # convert to real space
    potential=fftn(potential)
    return potential

#oxygen
potentials[8]=getAtomPotentialBox(atomBox, 8)
#Strontium
potentials[38]=getAtomPotentialBox(atomBox, 38)
#Titanium
potentials[22]=getAtomPotentialBox(atomBox, 22)
```

```python
#mask the atom potential boxes outside of the allowed radius
#  This is important to avoid box artifacts in the sample potential
z,y,x = np.ogrid[-pts/2:pts/2, -pts/2:pts/2, -pts/2:pts/2]
mask = x*x + y*y + z*z<= (pts/2)*(pts/2)

for potential in potentials:
    potential = potential*mask
```

```python
fix, axs=plt.subplots(1,3)
img=axs[0].imshow(potentials[8].real.sum(axis=2), extent=[-radius,radius,-radius,radius])
axs[0].set_xlabel("Angstrom")
axs[0].set_ylabel("Angstrom")
axs[0].set_title("Oxygen")
img.set_clim(potentials[38].real.sum(axis=2).min(),potentials[38].real.sum(axis=2).max())
img=axs[1].imshow(potentials[22].real.sum(axis=2), extent=[-radius,radius,-radius,radius])
axs[1].set_xlabel("Angstrom")
axs[1].set_ylabel("Angstrom")
axs[1].set_title("Titanium")
img.set_clim(potentials[38].real.sum(axis=2).min(),potentials[38].real.sum(axis=2).max())
axs[2].imshow(potentials[38].real.sum(axis=2), extent=[-radius,radius,-radius,radius])
axs[2].set_xlabel("Angstrom")
axs[2].set_ylabel("Angstrom")
axs[2].set_title("Strontium")
```

# Propagation through slices (slow)

### (This is a good place to parallelize using a GPU)

Propagation:

$$ \psi_{n+1}(x,y) = {p_n(k_x, k_y, \Delta z_n)\otimes[t_n(x,y)\psi_n(x,y)]} $$

In English: the wavefunction at the next slice is the convolution of the propagator function with the sample potential times the probe wavefunction.

In practice, it is best to speed up convolution by doing it as multiplication in Fourier space:

$$ \psi_{n+1}(x,y) = FT^{-1}(P_n(k_x, k_y, \Delta z_n)*FT[t_n(x,y)\psi_n(x,y)]) $$



First: the propagators:
========================

$$ P_n(k_x, k_y, \Delta z_n) = exp^{-i\pi\lambda (k_x^2+k_y^2) \Delta z}$$

```python
k = n/2.0*d
k2 = np.linspace(-k,k,n)**2
propx, propy = np.meshgrid(k2,k2)

prop = np.exp(0-1j*np.pi*p.wavelength*(propx+propy)*slice_thickness)

fig, axs = plt.subplots(1,2)
axs[0].imshow(prop.real, extent=[-k,k,-k,k], )
axs[0].set_xlabel("1/Angstrom")
axs[0].set_ylabel("1/Angstrom")
axs[0].set_title("Propagator Amplitude")
axs[1].imshow(prop.imag, extent=[-k,k,-k,k])
axs[1].set_xlabel("Angstrom")
axs[1].set_ylabel("Angstrom")
axs[1].set_title("Propagator Phase")
```

## Form the diffraction pattern

```python
diffpat = np.array(wave.real**2+wave.imag**2,dtype=np.float)

#TODO: translate k-space into scattering angle (mrad)
angle = k*p.wavelength*1000
plt.imshow(diffpat, extent=[-angle,angle,-angle,angle], )
plt.xlabel("mrad")
plt.ylabel("mrad")
plt.title("Diffraction pattern")
```

## Apply a mask to select the area of the detector

```python
k = n/2*d
k2 = np.linspace(-k,k,n)**2
xgrid, ygrid = np.meshgrid(k2,k2)
mask = np.ones_like(diffpat)

# detector angles in mrad
inner_radius = 5
outer_radius = 10

#convert angles to k-space
inner_k = inner_radius/1000.0/p.wavelength
outer_k = outer_radius/1000.0/p.wavelength

# values outside the outer radius are set to zero
mask = mask*((xgrid+ygrid)<outer_k**2)
# values inside the inner radius are set to zero
mask = mask*((xgrid+ygrid)>inner_k**2)
# mask is multiplied by data
masked_diffpat = diffpat * mask

# diffraction patterns:
angle = k*p.wavelength*1000
fig, axs = plt.subplots(1,2)
axs[0].imshow(diffpat, extent=[-angle,angle,-angle,angle], )
axs[0].set_xlabel("mrad")
axs[0].set_ylabel("mrad")
axs[0].set_title("Diffraction pattern")
axs[1].imshow(masked_diffpat, extent=[-angle,angle,-angle,angle])
axs[1].set_xlabel("mrad")
axs[1].set_ylabel("mrad")
axs[1].set_title("Diffraction Pattern falling on detector")
```

## Sum the intensity from that area

```python
pixel_intensity = np.sum(masked_diffpat)
print pixel_intensity
```

```python
#normalize pixel intensity to fraction of wavefunction's intensity
pixel_intensity = pixel_intensity/(n*n)
print pixel_intensity
```
